#Preparing Data
1. Copy the function 'df=pd.read_csv'
2. Copy in '('gdrive/My Drive/NAME OF YOUR CSV)', using the name of your csv file
3. Copy df.head(), this function will show some of your data, if your data set is very large it will only show a portion of it

```python
df=pd.read_csv('gdrive/My Drive/CountyHealthData_2014-2015.csv')
df.head()
```

#Creating a Data Subsection

1. Copy 'df[df["State"]== "NC"]' to show just data from North Carolina

If you want to look at another state, simply put that state's abbreviation into the function. If you want to look at Region, Division, or County, simply put that value where it says "State".

```python
df[df["State"]== "NC"]
```

#Defining
Now the North Carolina subsection needs to be defined. This will allow us to export this subsection as a csv.
1. The subset will be called NC_subset
2. Attach this name with an equal sign to 'df[df["State"]== "NC"].copy()'


```python
NC_subset = df[df["State"]== "NC"].copy()
```

3. NC_subset will show the data

```python
NC_subset
```

#Exporting
 1. To export the subset as a CSV use the function 'NC_subset.to_csv("NC_subset.csv", index=False)'

```python
NC_subset.to_csv("NC_subset.csv", index=False)
```

**The subset is now downloaded as a csv!**
